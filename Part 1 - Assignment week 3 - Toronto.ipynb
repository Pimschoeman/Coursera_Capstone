<h5> Importing the pandas library</h5>

```python
import pandas as pd
```

<h5> Reading the postal codes into a pandas dataframe</h5>

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
df.head(10)
```

<h5> Creating a new dataframe without the "Not assigned" values </h5>

```python
# i assumed that i could make a new dataframe with specifying that "not assigned" would not appear in column borough. 
dfna= df[df.Borough != 'Not assigned']
```

```python
dfna.head(10)
```

```python
#first i group the dfwna by postcode and borough, then i group the neighbourhoods per Borough in one column. 
dfna =dfna.groupby(['Postcode','Borough'])
dfna=dfna['Neighbourhood'].apply(', '.join).reset_index() 
dfna.head(10)
```

<h5> Iterating through the dataframe and assigning values of boroughs to neighbourhoods if they are not assigned. </h5>

```python
for i in range(len(dfna)):
    if dfna['Neighbourhood'].iloc[i] =='Not assigned':
        dfna['Neighbourhood'].iloc[i]= dfna['Borough'].iloc[i]
        
dfna
```
